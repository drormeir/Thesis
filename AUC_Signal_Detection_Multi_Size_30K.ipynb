```python
from Higher_Criticism import Bonferroni, Benjamini_Hochberg, Higher_Criticism
from Higher_Analysis import AUC_asymptotic_analysis_multi_size
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'monospace'
```

```python
gamma_power = np.linspace(0.1,0.9,9)
gamma_factor = np.linspace(0.1,0.4,4)
gamma_range = np.hstack([gamma_factor, -gamma_power])
N_range = [1000, 2000, 3000, 5000, 6000, 7000, 8000, 10000, 20000, 30000]
hc_models = [Bonferroni(), Benjamini_Hochberg(alpha=0.5)]
hc_models += [Higher_Criticism(gamma=gamma_range)]
monte_carlo = 10000
chunk_size = 100
```

```python

AUC_asymptotic_analysis_multi_size(N_range=N_range, beta_range=[0.7], r_range=[0.1, 0.2, 0.3], hc_models=hc_models, monte_carlo=monte_carlo, chunk_size=chunk_size)
AUC_asymptotic_analysis_multi_size(N_range=N_range, beta_range=[0.5], r_range=[0.05], hc_models=hc_models, monte_carlo=monte_carlo, chunk_size=chunk_size)
```
