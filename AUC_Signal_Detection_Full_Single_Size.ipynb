```python
from Higher_Criticism import Higher_Criticism, Benjamini_Hochberg, Bonferroni, Lowest_Angle, Berk_Jones, Base_Rejection_Method
from Higher_Analysis import AUC_full_analysis_single_size
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'monospace'
```

```python
N=30000
monte_carlo = 1000
chunk_size = 100
beta_range=list(np.arange(0.5,1.0,0.1))
r_range=list(np.arange(0.05,0.8,0.05))
gamma = list(np.arange(0.1,1.0,0.1))
gamma_power = list(-np.arange(0.1,1.0,0.1))
gamma_range = gamma + gamma_power

major_models = [Higher_Criticism(gamma=gamma_range, selection_method=Base_Rejection_Method.Selection_Method_ArgMin),\
                Benjamini_Hochberg(alpha=0.5, gamma=gamma_range, selection_method=Base_Rejection_Method.Selection_Method_ArgMin),\
                Benjamini_Hochberg(alpha=0.1, gamma=gamma_range, selection_method=Base_Rejection_Method.Selection_Method_ArgMin),\
                Bonferroni(alpha=1, gamma=gamma_range, selection_method=Base_Rejection_Method.Selection_Method_ArgMin),\
                Lowest_Angle(gamma=gamma_range, selection_method=Base_Rejection_Method.Selection_Method_ArgMin),\
                Berk_Jones(gamma=gamma_range, selection_method=Base_Rejection_Method.Selection_Method_ArgMin)]
```

```python

AUC_full_analysis_single_size(N=N, beta_range=beta_range, r_range=r_range,\
                              major_models=major_models,\
                              monte_carlo=monte_carlo, chunk_size=chunk_size)
```
